```python
import pandas as pd
import numpy as np
```

```python
events_df = pd.read_csv("../data/events.csv")
events_df.head()
```

```python
grouped_events = pd.pivot_table(events_df, values='minuto', index=['id_partido', 'id_equipo'], columns=['evento'], aggfunc='count', fill_value=0)
```

```python
grouped_events.reset_index()
```

```python
training = pd.read_csv('../data/train_matches.csv')
## droping players ids
training = training[['id_partido', 'id_equipo_local', 'id_equipo_visitante', 'ganador']]
training.head()
```

```python
no_events = training
only_local_events = no_events.merge(grouped_events, how='left', left_on=['id_partido', 'id_equipo_local'], right_on=['id_partido', 'id_equipo'])
training_with_events = only_local_events.merge(grouped_events, suffixes=("_local", "_visitante"),how='left', left_on=['id_partido', 'id_equipo_visitante'], right_on=['id_partido', 'id_equipo'])

training_with_events.head()
```

```python
## making ganador last column

cols = list(training_with_events.columns)
cols = list(filter(lambda name: name != "ganador", cols)) + ['ganador']
training_with_events = training_with_events[cols]
training_with_events.head()
```

```python
training_with_events.to_csv('basic_match_events_training.csv')
```
